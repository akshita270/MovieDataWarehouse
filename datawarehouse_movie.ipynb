```python
import pandas as pd
df_movies = pd.read_csv("movies.csv")
df_ratings = pd.read_csv("ratings.csv")
df_tags = pd.read_csv("tags.csv")
df_links = pd.read_csv("links.csv")
```

### Data transformation on movie dataset

```python
# Convert timestamp columns to datetime
df_movies['timestamp'] = pd.to_datetime(df_movies['timestamp'])

```

```python
df_movies
```

```python
# Extract year from movie titles and handle NaN values
df_movies['year'] = df_movies['title'].str.extract(r'\((\d{4})\)', expand=False)
```

```python
df_movies
```

```python
# Fill NaNs with a default value like 0
df_movies['year'] = df_movies['year'].fillna(0).astype(int) 
```

```python
# Convert to integer after handling NaN
df_movies['year'] = df_movies['year'].astype(int)
df_movies['year']
```

```python
# Convert title to string and fill NaN with a default value
df_movies['title'] = df_movies['title'].astype(str).fillna("Unknown")
df_movies['title']
```

```python
df_movies
```

```python
# Normalize movie titles 
df_movies['title'] = df_movies['title'].str.lower()
df_movies['title']
```

```python
df_movies
```

```python
df_movies['genres'] = df_movies['genres'].str.split('|')
df_movies['genres']
```

```python
df_movies
```

```python
# Check for missing values
print(df_movies.isna().sum())  
```

### Data Transformation on rating dataset

```python
df_ratings
```

```python
# Convert timestamp to datetime
df_ratings['timestamp'] = pd.to_datetime(df_ratings['timestamp'])
df_ratings['timestamp']
```

```python
# Check for missing values 
missing_values = df_ratings.isna().sum()
print(missing_values) 
```

```python
# Drop rows with missing data
df_ratings = df_ratings.dropna(subset=['userId', 'movieId', 'rating'])  
```

```python
# Normalize ratings
df_ratings['rating'] = df_ratings['rating'].astype(float).round(1)  

# Normalize user IDs and movie IDs
df_ratings['userId'] = df_ratings['userId'].astype(int)  
df_ratings['movieId'] = df_ratings['movieId'].astype(int)
```

```python
# Extract year from timestamp 
df_ratings['year'] = df_ratings['timestamp'].dt.year 
df_ratings['year']
```

```python
# Calculate average rating for each movie
df_avg_ratings = df_ratings.groupby('movieId')['rating'].mean().reset_index(name='avg_rating')

# Calculate the number of ratings per movie
df_rating_counts = df_ratings.groupby('movieId')['userId'].count().reset_index(name='rating_count')
```

```python
df_avg_ratings 
```

```python
df_rating_counts
```

```python
# Check for unique IDs
assert df_ratings['userId'].is_unique == False, "User IDs should not be unique"
assert df_ratings['movieId'].is_unique == False, "Movie IDs should not be unique"
```

```python
conn = sqlite3.connect("datawarehouse.db")
cursor = conn.cursor()
cursor.execute("PRAGMA foreign_keys = ON")

cursor.execute("""
CREATE TABLE fact_ratings (
  rating_uid INTEGER PRIMARY KEY,
  userId INTEGER,
  movieId INTEGER,
  rating DECIMAL(3, 1),
  timestamp DATETIME,
  FOREIGN KEY(userId) REFERENCES dim_users(userId),
  FOREIGN KEY(movieId) REFERENCES dim_movies(movieId)
);
""")
```

```python
cursor.execute("""
CREATE TABLE dimension_users (
  userId INTEGER PRIMARY KEY,
  name VARCHAR(255),
  age INTEGER,
  gender CHAR(1)
);
""")
```

```python
cursor.execute("""
CREATE TABLE dimension_movies (
  movieId INTEGER  PRIMARY KEY,
  title VARCHAR(255),
  genres VARCHAR(255),
  year INTEGER,
  FOREIGN KEY(movieId) REFERENCES fact_ratings(movieId) 
);
""")
```

```python
cursor.execute("""
CREATE TABLE dimension_tags (
  tag_id INTEGER PRIMARY KEY,
  userId INTEGER,
  movieId INTEGER,
  tag VARCHAR(255),
  FOREIGN KEY(movieId) REFERENCES dimension_movies(movieId),
  FOREIGN KEY(userId) REFERENCES dimension_users(userId)
);
""")
```

```python
cursor.execute("""
CREATE TABLE dimension_links (
  movieId INTEGER PRIMARY KEY,
  imdbId INTEGER,  
  tmdbId INTEGER, 
  FOREIGN KEY(movieId) REFERENCES dimension_movies(movieId) 
);
""")
```

```python
conn.close()
```

```python
conn_source = sqlite3.connect('Movies.db')  
cursor_source = conn_source.cursor()

conn_target = sqlite3.connect('datawarehouse.db') 
cursor_target = conn_target.cursor()

# Fetch data from the source movies table
select_query = """
SELECT movieId, title, genres, year  
FROM movies;  
"""

# Execute the query and fetch the data
cursor_source.execute(select_query)
movies_data = cursor_source.fetchall()  

# for inserting data into the dimension_movies table
insert_query = """
INSERT INTO dimension_movies (movieId, title, genres, year)
VALUES (?, ?, ?, ?);
"""

# Insert each row from source to target
for row in movies_data:
    cursor_target.execute(insert_query, row) 


conn_target.commit() 
conn_source.close()  
conn_target.close()  
```

```python
import sqlite3

def sync_Movies_to_data_warehouse():
    try:
        # Connect to the Movies and data warehouse databases
        conn_Movies = sqlite3.connect('Movies.db')
        conn_data_warehouse = sqlite3.connect('datawarehouse.db')

        cursor_Movies = conn_Movies.cursor()
        cursor_data_warehouse = conn_data_warehouse.cursor()

        # Fetch data from the "movies" table from Movies database to insert into the datawarehouse
        cursor_Movies.execute("SELECT movieId, title, genres, year FROM movies")
        movies_data = cursor_Movies.fetchall()

       
        insert_query_movies = """
        INSERT OR REPLACE INTO dimension_movies (movieId, title, genres, year)
        VALUES (?, ?, ?, ?);
        """

        # Insert each row from the Movies to the data warehouse
        for row in movies_data:
            cursor_data_warehouse.execute(insert_query_movies, row)

       
        conn_data_warehouse.commit()

        # Synchronize data from Movies to dimension_movies
        cursor_Movies.execute("SELECT  movieId, title, genres ,year from movies")
        movies_data = cursor_Movies.fetchall()

        insert_query_movies = """
        INSERT OR REPLACE INTO dimension_movies (movieId, title, genres, year)
        VALUES (?, ?, ?, ?);
        """

        for row in movies_data:
            cursor_data_warehouse.execute(insert_query_movies, row)

        
        conn_data_warehouse.commit()

        print("Data synchronization completed successfully!")

    except sqlite3.Error as e:
        print(f"Error occurred: {e}")
        print("Data that caused the error:")
        print(movies_data)  

    finally:
      
        conn_Movies.close()
        conn_data_warehouse.close()

sync_Movies_to_data_warehouse()
```
